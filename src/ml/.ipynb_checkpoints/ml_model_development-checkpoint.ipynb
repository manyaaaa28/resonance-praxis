## Customer Behavior Prediction Models

This notebook develops machine learning models for predicting customer behavior using the shopping trends dataset.

**Models Implemented:**
1. Purchase Probability Predictor
2. Revenue Forecasting Model  
3. Customer Churn Predictor
4. Price Elasticity Calculator

## 1. Data Loading & Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load the shopping trends dataset
df = pd.read_csv('shopping_trends.csv')

print(f"Dataset Shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
df.head()
```

```python
# Data info and statistics
print("=" * 50)
print("DATASET SUMMARY")
print("=" * 50)
df.info()
print("\n" + "=" * 50)
print("NUMERICAL STATISTICS")
print("=" * 50)
df.describe()
```

## 2. Feature Engineering

```python
def engineer_features(data):
    """
    Create behavioral scoring features from raw customer data
    """
    df_features = data.copy()
    
    # 1. Price Sensitivity Score
    df_features['price_sensitivity'] = 0.0
    df_features.loc[df_features['Purchase Amount (USD)'] < 40, 'price_sensitivity'] += 0.3
    df_features.loc[(df_features['Purchase Amount (USD)'] >= 40) & (df_features['Purchase Amount (USD)'] < 60), 'price_sensitivity'] += 0.15
    df_features.loc[df_features['Discount Applied'] == 'Yes', 'price_sensitivity'] += 0.2
    df_features.loc[df_features['Promo Code Used'] == 'Yes', 'price_sensitivity'] += 0.2
    df_features.loc[df_features['Shipping Type'] == 'Free Shipping', 'price_sensitivity'] += 0.15
    df_features.loc[df_features['Shipping Type'] == 'Standard', 'price_sensitivity'] += 0.1
    df_features['price_sensitivity'] = df_features['price_sensitivity'].clip(0, 1)
    
    # 2. Loyalty Score
    df_features['loyalty_score'] = 0.0
    df_features.loc[df_features['Previous Purchases'] > 40, 'loyalty_score'] += 0.35
    df_features.loc[(df_features['Previous Purchases'] > 25) & (df_features['Previous Purchases'] <= 40), 'loyalty_score'] += 0.2
    df_features.loc[(df_features['Previous Purchases'] > 10) & (df_features['Previous Purchases'] <= 25), 'loyalty_score'] += 0.1
    df_features.loc[df_features['Subscription Status'] == 'Yes', 'loyalty_score'] += 0.25
    df_features.loc[df_features['Review Rating'] >= 4.5, 'loyalty_score'] += 0.2
    df_features.loc[(df_features['Review Rating'] >= 4.0) & (df_features['Review Rating'] < 4.5), 'loyalty_score'] += 0.1
    df_features['loyalty_score'] = df_features['loyalty_score'].clip(0, 1)
    
    # 3. Impulse Score
    df_features['impulse_score'] = 0.0
    df_features.loc[df_features['Shipping Type'] == 'Next Day Air', 'impulse_score'] += 0.3
    df_features.loc[df_features['Shipping Type'] == 'Express', 'impulse_score'] += 0.2
    df_features.loc[df_features['Age'] < 30, 'impulse_score'] += 0.2
    df_features.loc[(df_features['Age'] >= 30) & (df_features['Age'] < 40), 'impulse_score'] += 0.1
    df_features.loc[df_features['Category'] == 'Accessories', 'impulse_score'] += 0.15
    df_features['impulse_score'] = df_features['impulse_score'].clip(0, 1)
    
    # 4. Quality Preference Score
    df_features['quality_preference'] = 0.0
    df_features.loc[df_features['Purchase Amount (USD)'] > 80, 'quality_preference'] += 0.3
    df_features.loc[(df_features['Purchase Amount (USD)'] > 60) & (df_features['Purchase Amount (USD)'] <= 80), 'quality_preference'] += 0.15
    df_features.loc[df_features['Review Rating'] >= 4.5, 'quality_preference'] += 0.15
    df_features.loc[df_features['Age'] > 45, 'quality_preference'] += 0.1
    df_features.loc[df_features['Discount Applied'] == 'No', 'quality_preference'] += 0.1
    df_features['quality_preference'] = df_features['quality_preference'].clip(0, 1)
    
    return df_features

# Apply feature engineering
df_engineered = engineer_features(df)
print("✅ Feature Engineering Complete!")
print(f"\nNew Features Created:")
print("- price_sensitivity")
print("- loyalty_score")
print("- impulse_score")
print("- quality_preference")

df_engineered[['Customer ID', 'price_sensitivity', 'loyalty_score', 'impulse_score', 'quality_preference']].head(10)
```

## 3. Customer Segmentation (Clustering)

```python
# Segment customers into personas based on behavioral scores
def segment_customers(df):
    segments = {}
    
    segments['budget'] = df[df['price_sensitivity'] > 0.6].copy()
    segments['premium'] = df[df['quality_preference'] > 0.6].copy()
    segments['impulse'] = df[df['impulse_score'] > 0.6].copy()
    segments['loyal'] = df[df['loyalty_score'] > 0.6].copy()
    
    return segments

customer_segments = segment_customers(df_engineered)

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
colors = ['#f59e0b', '#6366f1', '#10b981', '#ec4899']
segment_names = ['Budget Hunter', 'Premium Buyer', 'Impulse Shopper', 'Loyal Customer']

for idx, (segment_key, segment_data) in enumerate(customer_segments.items()):
    row = idx // 2
    col = idx % 2
    
    axes[row, col].hist(segment_data['Purchase Amount (USD)'], bins=30, color=colors[idx], alpha=0.7, edgecolor='black')
    axes[row, col].set_title(f"{segment_names[idx]} ({len(segment_data)} customers)", fontsize=14, fontweight='bold')
    axes[row, col].set_xlabel('Purchase Amount (USD)')
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].axvline(segment_data['Purchase Amount (USD)'].mean(), color='red', linestyle='--', linewidth=2, label=f"Avg: ${segment_data['Purchase Amount (USD)'].mean():.2f}")
    axes[row, col].legend()

plt.tight_layout()
plt.savefig('images/customer_segments.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n📊 CUSTOMER SEGMENTATION RESULTS:")
print("=" * 60)
for segment_key, segment_data in customer_segments.items():
    print(f"{segment_key.upper()}: {len(segment_data)} customers (Avg Purchase: ${segment_data['Purchase Amount (USD)'].mean():.2f})")
```

## 4. Model 1: Purchase Probability Predictor

```python
def predict_purchase_probability(customer, scenario, weights=None):
    """
    Predict probability of purchase using weighted feature scoring
    """
    if weights is None:
        weights = {
            'price_sensitivity': 0.25,
            'loyalty_score': 0.20,
            'category_match': 0.15,
            'discount_impact': 0.20,
            'seasonal_fit': 0.10,
            'affordability': 0.10
        }
    
    probability = 0.0
    
    # Price sensitivity impact
    discount = scenario.get('discount', 0)
    if customer['price_sensitivity'] > 0.6:
        price_impact = min(1, 0.3 + (discount / 100) * 0.7)
    else:
        price_impact = 0.7 + (discount / 100) * 0.2
    probability += price_impact * weights['price_sensitivity']
    
    # Loyalty impact
    probability += customer['loyalty_score'] * weights['loyalty_score']
    
    # Category match
    category_match = 1 if customer['Category'] == scenario.get('category', 'Clothing') else 0.3
    probability += category_match * weights['category_match']
    
    # Discount effectiveness
    discount_impact = min(1, 0.4 + (discount / 50) * customer['price_sensitivity'])
    probability += discount_impact * weights['discount_impact']
    
    # Seasonal fit
    seasonal_match = 1 if customer['Season'] == scenario.get('season', 'Spring') else 0.5
    probability += seasonal_match * weights['seasonal_fit']
    
    # Affordability
    base_price = scenario.get('basePrice', 75)
    final_price = base_price * (1 - discount / 100)
    affordability = min(1, customer['Purchase Amount (USD)'] / final_price)
    probability += affordability * weights['affordability']
    
    return min(0.95, max(0.05, probability))

# Test the model
test_scenario = {'discount': 25, 'category': 'Clothing', 'season': 'Spring', 'basePrice': 75}
sample_predictions = []

for idx in range(min(100, len(df_engineered))):
    customer = df_engineered.iloc[idx]
    prob = predict_purchase_probability(customer, test_scenario)
    sample_predictions.append(prob)

# Visualize predictions
plt.figure(figsize=(12, 6))
plt.hist(sample_predictions, bins=20, color='#6366f1', alpha=0.7, edgecolor='black')
plt.axvline(np.mean(sample_predictions), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(sample_predictions):.3f}')
plt.title('Purchase Probability Distribution (Sample of 100 customers)', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)
plt.savefig('images/purchase_probability_dist.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"✅ Purchase Probability Model Tested")
print(f"Mean Probability: {np.mean(sample_predictions):.3f}")
print(f"Std Dev: {np.std(sample_predictions):.3f}")
```

## 5. Model 2: Revenue Forecasting

```python
def forecast_revenue(customers, scenario):
    """
    Forecast revenue using purchase probability model
    """
    base_price = scenario.get('basePrice', 75)
    discount = scenario.get('discount', 0)
    final_price = base_price * (1 - discount / 100)
    
    total_revenue = 0
    expected_purchases = 0
    
    for idx in range(len(customers)):
        customer = customers.iloc[idx]
        prob = predict_purchase_probability(customer, scenario)
        
        # Expected value calculation
        expected_value = prob * final_price
        total_revenue += expected_value
        
        if prob > 0.5:
            expected_purchases += 1
    
    conversion_rate = (expected_purchases / len(customers)) * 100
    avg_order_value = total_revenue / expected_purchases if expected_purchases > 0 else 0
    
    return {
        'total_revenue': round(total_revenue),
        'expected_purchases': round(expected_purchases),
        'conversion_rate': round(conversion_rate, 2),
        'avg_order_value': round(avg_order_value, 2)
    }

# Test different discount scenarios
discount_levels = [0, 10, 20, 30, 40, 50]
revenue_forecasts = []

for discount in discount_levels:
    scenario = {'discount': discount, 'category': 'Clothing', 'season': 'Spring', 'basePrice': 75}
    forecast = forecast_revenue(df_engineered, scenario)
    forecast['discount'] = discount
    revenue_forecasts.append(forecast)

# Create DataFrame for visualization
forecast_df = pd.DataFrame(revenue_forecasts)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Revenue vs Discount
axes[0].plot(forecast_df['discount'], forecast_df['total_revenue'], marker='o', linewidth=2, markersize=8, color='#10b981')
axes[0].set_title('Revenue Forecast vs Discount Level', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Discount (%)')
axes[0].set_ylabel('Total Revenue ($)')
axes[0].grid(alpha=0.3)

# Conversion Rate vs Discount
axes[1].plot(forecast_df['discount'], forecast_df['conversion_rate'], marker='s', linewidth=2, markersize=8, color='#6366f1')
axes[1].set_title('Conversion Rate vs Discount Level', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Discount (%)')
axes[1].set_ylabel('Conversion Rate (%)')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('images/revenue_forecast.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n📈 REVENUE FORECASTING RESULTS:")
print("=" * 70)
print(forecast_df.to_string(index=False))

# Find optimal discount
optimal = forecast_df.loc[forecast_df['total_revenue'].idxmax()]
print(f"\n🎯 OPTIMAL DISCOUNT: {optimal['discount']}% (Revenue: ${optimal['total_revenue']:,})")
```

## 6. Model 3: Churn Prediction

```python
def predict_churn_risk(customer):
    """
    Predict customer churn risk based on behavioral signals
    """
    churn_score = 0.0
    risk_factors = []
    
    # Purchase frequency
    freq = customer.get('Frequency of Purchases', 'Monthly')
    if freq == 'Annually':
        churn_score += 0.3
        risk_factors.append('Low purchase frequency')
    elif freq in ['Quarterly', 'Every 3 Months']:
        churn_score += 0.15
    
    # Satisfaction (Review Rating)
    rating = customer.get('Review Rating', 3.5)
    if rating < 3.5:
        churn_score += 0.25
        risk_factors.append('Low satisfaction score')
    
    # Subscription status
    if customer.get('Subscription Status', 'No') == 'No' and customer['loyalty_score'] > 0.5:
        churn_score += 0.2
        risk_factors.append('No subscription (loyalty risk)')
    
    # Previous purchases
    if customer.get('Previous Purchases', 0) < 5:
        churn_score += 0.15
        risk_factors.append('New customer')
    
    churn_score = min(1, churn_score)
    risk_level = 'HIGH' if churn_score > 0.6 else 'MEDIUM' if churn_score > 0.3 else 'LOW'
    
    return {
        'churn_score': round(churn_score, 3),
        'risk_level': risk_level,
        'risk_factors': risk_factors
    }

# Predict churn for all customers
churn_predictions = []
for idx in range(len(df_engineered)):
    customer = df_engineered.iloc[idx]
    prediction = predict_churn_risk(customer)
    prediction['customer_id'] = customer['Customer ID']
    churn_predictions.append(prediction)

churn_df = pd.DataFrame(churn_predictions)

# Visualize churn risk distribution
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Risk level counts
risk_counts = churn_df['risk_level'].value_counts()
colors_risk = {'HIGH': '#ef4444', 'MEDIUM': '#f59e0b', 'LOW': '#10b981'}
axes[0].bar(risk_counts.index, risk_counts.values, color=[colors_risk[x] for x in risk_counts.index])
axes[0].set_title('Customer Churn Risk Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Risk Level')
axes[0].set_ylabel('Number of Customers')
axes[0].grid(alpha=0.3, axis='y')

# Churn score distribution
axes[1].hist(churn_df['churn_score'], bins=20, color='#ec4899', alpha=0.7, edgecolor='black')
axes[1].axvline(0.3, color='orange', linestyle='--', linewidth=2, label='Medium Risk Threshold')
axes[1].axvline(0.6, color='red', linestyle='--', linewidth=2, label='High Risk Threshold')
axes[1].set_title('Churn Score Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Churn Score')
axes[1].set_ylabel('Frequency')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('images/churn_risk_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# Show high-risk customers
high_risk = churn_df[churn_df['risk_level'] == 'HIGH'].sort_values('churn_score', ascending=False).head(10)
print("\n⚠️ TOP 10 HIGH-RISK CUSTOMERS:")
print("=" * 60)
print(high_risk[['customer_id', 'churn_score', 'risk_level']].to_string(index=False))
```

## 7. Model 4: Price Elasticity

```python
def calculate_price_elasticity(customers, base_scenario):
    """
    Calculate price elasticity of demand
    """
    base_discount = base_scenario.get('discount', 20)
    
    # Test different discount levels
    test_points = [
        {'discount': base_discount - 10, 'label': 'Lower'},
        {'discount': base_discount, 'label': 'Current'},
        {'discount': base_discount + 10, 'label': 'Higher'}
    ]
    
    results = []
    for point in test_points:
        scenario = {**base_scenario, 'discount': max(0, min(100, point['discount']))}
        forecast = forecast_revenue(customers, scenario)
        results.append({
            'discount': point['discount'],
            'label': point['label'],
            'conversion_rate': forecast['conversion_rate'],
            'revenue': forecast['total_revenue'],
            'purchases': forecast['expected_purchases']
        })
    
    # Calculate elasticity
    current = results[1]
    higher = results[2]
    
    price_change = ((100 - higher['discount']) - (100 - current['discount'])) / (100 - current['discount']) * 100
    demand_change = (higher['purchases'] - current['purchases']) / current['purchases'] * 100
    
    elasticity = demand_change / price_change if price_change != 0 else 0
    
    # Find optimal
    optimal = max(results, key=lambda x: x['revenue'])
    
    return {
        'elasticity': round(elasticity, 2),
        'interpretation': 'Elastic (price-sensitive)' if abs(elasticity) > 1 else 'Inelastic',
        'results': results,
        'optimal_discount': optimal['discount'],
        'optimal_revenue': optimal['revenue']
    }

# Calculate elasticity
base_scenario = {'discount': 20, 'category': 'Clothing', 'season': 'Spring', 'basePrice': 75}
elasticity_analysis = calculate_price_elasticity(df_engineered, base_scenario)

# Visualization
results_df = pd.DataFrame(elasticity_analysis['results'])

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(results_df['discount'], results_df['revenue'], marker='o', linewidth=3, markersize=10, color='#6366f1')
ax.set_title(f"Price Elasticity Analysis (Elasticity Coefficient: {elasticity_analysis['elasticity']})", fontsize=14, fontweight='bold')
ax.set_xlabel('Discount (%)', fontsize=12)
ax.set_ylabel('Revenue ($)', fontsize=12)
ax.grid(alpha=0.3)

# Mark optimal point
optimal_idx = results_df[results_df['discount'] == elasticity_analysis['optimal_discount']].index[0]
ax.scatter(elasticity_analysis['optimal_discount'], elasticity_analysis['optimal_revenue'], 
           s=200, color='red', zorder=5, label=f"Optimal: {elasticity_analysis['optimal_discount']}% (${elasticity_analysis['optimal_revenue']:,})")
ax.legend(fontsize=11)

plt.savefig('images/price_elasticity.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n📊 PRICE ELASTICITY ANALYSIS:")
print("=" * 70)
print(f"Elasticity Coefficient: {elasticity_analysis['elasticity']}")
print(f"Interpretation: {elasticity_analysis['interpretation']}")
print(f"\n🎯 Optimal Discount: {elasticity_analysis['optimal_discount']}%")
print(f"Projected Revenue: ${elasticity_analysis['optimal_revenue']:,}")
```
